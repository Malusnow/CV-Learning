```python
import cv2 as cv
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Extract and visualize SIFT features
Let's look at the SIFT features extracted for a pair of images.

```python
# First, write the function for loading the images in a required format --- we will use it repeatedly.
# Change the path to the folder with the photos of white fox figurine
img_dir = r'C:\Users\wei\Documents\xwechat_files\wxid_vfo4erq7ri6922_23a5\msg\file\2025-07\07-08.data\07-08.data\white_fox_figurine'


def load_img(view_i):
    r"""Loads the image from {img_dir} for the viewpoint {view_i}.

    Parameters
    ----------
    view_i : int
        Viewpoint ID.

    Returns
    -------
    img : numpy.ndarray
        of shape [height, width] and type uint8, grayscale.
    """
    # Load the image, convert it to grayscale and make it numpy array, see the docscring of the function for more details.
    #
    # For loading the image and making it grayscale, see
    #   https://pillow.readthedocs.io/en/stable/reference/Image.html
    #   https://pillow.readthedocs.io/en/stable/reference/ImageOps.html
    # For conversion to numpy array, see
    #   https://numpy.org/doc/2.2/reference/generated/numpy.array.html
    img = f'{img_dir}/{view_i:04}.png'
    # 构建完整的图像文件路径
    img_path = os.path.join(img_dir, f'{view_i:04d}.png')
    
    # 使用PIL加载图像
    img = Image.open(img_path)
    
    # 使用ImageOps转换为灰度图
    img_gray = ImageOps.grayscale(img)
    
    # 转换为numpy数组
    img_array = np.array(img_gray, dtype=np.uint8)
    return img_array
```

```python
# Load the first image and crop it to the bounding box of the object for convenience
# Extract keypoints
# Draw the keypoints on top of the image
#
# For keypoint extraction and drawing, look for SIFT_create(), detect(), drawKeypoints() at
#   https://docs.opencv.org/4.11.0/d7/d60/classcv_1_1SIFT.html
#   https://docs.opencv.org/4.11.0/d4/d5d/group__features2d__draw.html
# 加载第一张图像并裁剪到目标物体的边界框
img1 = load_img(11)
img1 = np.array(img1[1100:1600, 900:1200])

# 创建SIFT检测器
sift = cv.SIFT_create()

# 提取关键点
keypoints1 = sift.detect(img1, None)

# 在图像上绘制关键点
img1_with_keypoints = cv.drawKeypoints(img1, keypoints1, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Load the second image, crop it, extract keypoints, draw
# 加载第二张图像，裁剪并提取关键点
img2 = load_img(7)
img2 = np.array(img2[1100:1600, 900:1200])

# 提取关键点
keypoints2 = sift.detect(img2, None)

# 绘制关键点
img2_with_keypoints = cv.drawKeypoints(img2, keypoints2, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show both images side by side for comparison, look for subplots(), imshow() at
#   https://matplotlib.org/3.10.3/api/pyplot_summary.html
# 并排显示两张图像进行比较
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 显示第一张图像
axes[0].imshow(img1_with_keypoints, cmap='gray')
axes[0].set_title(f'View 11 - Keypoints: {len(keypoints1)}')
axes[0].axis('off')

# 显示第二张图像
axes[1].imshow(img2_with_keypoints, cmap='gray')
axes[1].set_title(f'View 7 - Keypoints: {len(keypoints2)}')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print(f"Image 1 detected {len(keypoints1)} keypoints")
print(f"Image 2 detected {len(keypoints2)} keypoints")
```

## Robustness to distortions
Let's try extracting the SIFT features for distorted versions of the second image.

### Scale

```python
# Load the second image, crop it, resize it, extract keypoints, draw the first and the distorted second image
#
# Look for resize() at https://pillow.readthedocs.io/en/stable/reference/Image.html
# Load the second image, crop it, resize it, extract keypoints, draw the first and the distorted second image
#
# Look for resize() at https://pillow.readthedocs.io/en/stable/reference/Image.html
# 加载第二张图像并裁剪
img2 = load_img(7)
img2_cropped = np.array(img2[1100:1600, 900:1200])

# 创建SIFT检测器
sift = cv.SIFT_create()

# 提取原始图像的关键点
keypoints2_original = sift.detect(img2_cropped, None)
img2_original_with_keypoints = cv.drawKeypoints(img2_cropped, keypoints2_original, None, 
                                                flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 对第二张图像进行不同的变形处理
distortions = [
    ("Scale 0.5x", lambda img: img.resize((int(img.width * 0.5), int(img.height * 0.5)), Image.LANCZOS)),
    ("Scale 1.5x", lambda img: img.resize((int(img.width * 1.5), int(img.height * 1.5)), Image.LANCZOS)),
    ("Rotate 30°", lambda img: img.rotate(30, expand=True)),
    ("Rotate -45°", lambda img: img.rotate(-45, expand=True)),
]

# 创建子图显示结果
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 显示原始图像
axes[0, 0].imshow(img2_original_with_keypoints, cmap='gray')
axes[0, 0].set_title(f'Original Image - Keypoints: {len(keypoints2_original)}')
axes[0, 0].axis('off')

# 对每种变形进行处理
for i, (distortion_name, distortion_func) in enumerate(distortions):
    # 将numpy数组转换为PIL图像
    img_pil = Image.fromarray(img2_cropped)
    
    # 应用变形
    img_distorted_pil = distortion_func(img_pil)
    
    # 转换回numpy数组
    img_distorted = np.array(img_distorted_pil)
    
    # 提取变形图像的关键点
    keypoints_distorted = sift.detect(img_distorted, None)
    
    # 绘制关键点
    img_distorted_with_keypoints = cv.drawKeypoints(img_distorted, keypoints_distorted, None,
                                                    flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    
    # 计算子图位置
    row = (i + 1) // 3
    col = (i + 1) % 3
    
    # 显示变形后的图像
    axes[row, col].imshow(img_distorted_with_keypoints, cmap='gray')
    axes[row, col].set_title(f'{distortion_name} - Keypoints: {len(keypoints_distorted)}')
    axes[row, col].axis('off')
    
    print(f"{distortion_name}: Detected {len(keypoints_distorted)} keypoints")

# 隐藏空的子图
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()

print(f"\nOriginal image detected {len(keypoints2_original)} keypoints")
print("SIFT feature robustness test for image transformations completed")
```

### Rotation

```python
# Load the image, crop it, rotate, extract keypoints, draw the images
#
# Look for rotate() at https://pillow.readthedocs.io/en/stable/reference/Image.html
# 加载并裁剪图像
img = load_img(11)
img_cropped = np.array(img[1100:1600, 900:1200])

# 创建SIFT检测器
sift = cv.SIFT_create()

# 提取原始裁剪图像的关键点
keypoints_original = sift.detect(img_cropped, None)
img_original_with_keypoints = cv.drawKeypoints(img_cropped, keypoints_original, None, 
                                              flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 将numpy数组转换为PIL图像以便旋转
img_pil = Image.fromarray(img_cropped)

# 将图像旋转45度
img_rotated_pil = img_pil.rotate(45, expand=True)

# 转换回numpy数组
img_rotated = np.array(img_rotated_pil)

# 提取旋转图像的关键点
keypoints_rotated = sift.detect(img_rotated, None)
img_rotated_with_keypoints = cv.drawKeypoints(img_rotated, keypoints_rotated, None,
                                             flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 并排显示两张图像
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 显示原始图像
axes[0].imshow(img_original_with_keypoints, cmap='gray')
axes[0].set_title(f'Original Image - Keypoints: {len(keypoints_original)}')
axes[0].axis('off')

# 显示旋转图像
axes[1].imshow(img_rotated_with_keypoints, cmap='gray')
axes[1].set_title(f'Rotated 45° - Keypoints: {len(keypoints_rotated)}')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print(f"Original image detected {len(keypoints_original)} keypoints")
print(f"Rotated image detected {len(keypoints_rotated)} keypoints")
print("Image rotation and SIFT keypoint detection completed")
```

### Brightness

```python
# Load the image, crop it, make it brighter or darker, extract keypoints, draw the images
# 加载并裁剪图像
img = load_img(11)
img_cropped = np.array(img[1100:1600, 900:1200])

# 创建SIFT检测器
sift = cv.SIFT_create()

# 提取原始裁剪图像的关键点
keypoints_original = sift.detect(img_cropped, None)
img_original_with_keypoints = cv.drawKeypoints(img_cropped, keypoints_original, None, 
                                              flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 创建亮度调整的图像
# 使图像变亮 (增加像素值)
img_brighter = np.clip(img_cropped.astype(np.float32) * 1.5, 0, 255).astype(np.uint8)

# 使图像变暗 (减少像素值)
img_darker = np.clip(img_cropped.astype(np.float32) * 0.5, 0, 255).astype(np.uint8)

# 提取变亮图像的关键点
keypoints_brighter = sift.detect(img_brighter, None)
img_brighter_with_keypoints = cv.drawKeypoints(img_brighter, keypoints_brighter, None,
                                              flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 提取变暗图像的关键点
keypoints_darker = sift.detect(img_darker, None)
img_darker_with_keypoints = cv.drawKeypoints(img_darker, keypoints_darker, None,
                                            flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 显示三张图像
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 显示原始图像
axes[0].imshow(img_original_with_keypoints, cmap='gray')
axes[0].set_title(f'Original Image - Keypoints: {len(keypoints_original)}')
axes[0].axis('off')

# 显示变亮图像
axes[1].imshow(img_brighter_with_keypoints, cmap='gray')
axes[1].set_title(f'Brighter (1.5x) - Keypoints: {len(keypoints_brighter)}')
axes[1].axis('off')

# 显示变暗图像
axes[2].imshow(img_darker_with_keypoints, cmap='gray')
axes[2].set_title(f'Darker (0.5x) - Keypoints: {len(keypoints_darker)}')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print(f"Original image detected {len(keypoints_original)} keypoints")
print(f"Brighter image detected {len(keypoints_brighter)} keypoints")
print(f"Darker image detected {len(keypoints_darker)} keypoints")
print("Image brightness adjustment and SIFT keypoint detection completed")
```

### Noise

```python
# Load the image, crop it, add random noise, extract keypoints, draw the images
#
# See https://numpy.org/doc/2.2/reference/random/index.html
# 加载并裁剪图像
img = load_img(11)
img_cropped = np.array(img[1100:1600, 900:1200])

# 创建SIFT检测器
sift = cv.SIFT_create()

# 提取原始裁剪图像的关键点
keypoints_original = sift.detect(img_cropped, None)
img_original_with_keypoints = cv.drawKeypoints(img_cropped, keypoints_original, None, 
                                              flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 添加高斯噪声
gaussian_noise = np.random.normal(0, 20, img_cropped.shape).astype(np.float32)
img_gaussian_noise = np.clip(img_cropped.astype(np.float32) + gaussian_noise, 0, 255).astype(np.uint8)

# 添加椒盐噪声
img_salt_pepper = img_cropped.copy()
# 添加椒噪声 (黑点)
salt_pepper_prob = 0.02
salt_coords = np.random.random(img_cropped.shape) < salt_pepper_prob
img_salt_pepper[salt_coords] = 0
# 添加盐噪声 (白点)
pepper_coords = np.random.random(img_cropped.shape) < salt_pepper_prob
img_salt_pepper[pepper_coords] = 255

# 添加均匀噪声
uniform_noise = np.random.uniform(-30, 30, img_cropped.shape).astype(np.float32)
img_uniform_noise = np.clip(img_cropped.astype(np.float32) + uniform_noise, 0, 255).astype(np.uint8)

# 提取高斯噪声图像的关键点
keypoints_gaussian = sift.detect(img_gaussian_noise, None)
img_gaussian_with_keypoints = cv.drawKeypoints(img_gaussian_noise, keypoints_gaussian, None,
                                              flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 提取椒盐噪声图像的关键点
keypoints_salt_pepper = sift.detect(img_salt_pepper, None)
img_salt_pepper_with_keypoints = cv.drawKeypoints(img_salt_pepper, keypoints_salt_pepper, None,
                                                 flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 提取均匀噪声图像的关键点
keypoints_uniform = sift.detect(img_uniform_noise, None)
img_uniform_with_keypoints = cv.drawKeypoints(img_uniform_noise, keypoints_uniform, None,
                                             flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 显示四张图像
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 显示原始图像
axes[0, 0].imshow(img_original_with_keypoints, cmap='gray')
axes[0, 0].set_title(f'Original Image - Keypoints: {len(keypoints_original)}')
axes[0, 0].axis('off')

# 显示高斯噪声图像
axes[0, 1].imshow(img_gaussian_with_keypoints, cmap='gray')
axes[0, 1].set_title(f'Gaussian Noise - Keypoints: {len(keypoints_gaussian)}')
axes[0, 1].axis('off')

# 显示椒盐噪声图像
axes[1, 0].imshow(img_salt_pepper_with_keypoints, cmap='gray')
axes[1, 0].set_title(f'Salt & Pepper Noise - Keypoints: {len(keypoints_salt_pepper)}')
axes[1, 0].axis('off')

# 显示均匀噪声图像
axes[1, 1].imshow(img_uniform_with_keypoints, cmap='gray')
axes[1, 1].set_title(f'Uniform Noise - Keypoints: {len(keypoints_uniform)}')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

print(f"Original image detected {len(keypoints_original)} keypoints")
print(f"Gaussian noise image detected {len(keypoints_gaussian)} keypoints")
print(f"Salt & pepper noise image detected {len(keypoints_salt_pepper)} keypoints")
print(f"Uniform noise image detected {len(keypoints_uniform)} keypoints")
print("Image noise addition and SIFT keypoint detection completed")
```

## Finding corresponding features on different images
Let's try finding a pair of corresponding SIFT features on two different images.

```python
# Load the image and crop it, extract keypoints and their descriptors
# On the first image, select some keypoint that is easy to identify, e.g the one with the largest scale, or a keypoints by its coordinates
# Find the distance from this keypoint to each of the keypoints detected in the second image, in the descriptor space
# Select the keypoint from the second image with the minimal distance to the keypoint from the first image
# Draw the pair of images and the respective pair of keypoints and check if they represend the corresponding parts of the object
#
# For extraction of keypoint desriptors, look for compute() or detectAndCompute() at
#   https://docs.opencv.org/4.11.0/d7/d60/classcv_1_1SIFT.html
# For selecting the keypoint in the first image that is easy to identify (e.g by its size),
#  look at the attributes of the keypoints returned by detect() or detectAndCompute()
# You may need to find the index of the element of an array with the max/min value, see
#   https://numpy.org/doc/2.2/reference/generated/numpy.argmax.html
# To compute the distances between descriptors, see
#   https://numpy.org/doc/2.2/reference/generated/numpy.linalg.norm.html
# 加载并裁剪第一张图像
img1 = load_img(11)
img1_cropped = np.array(img1[1100:1600, 900:1200])

# 加载并裁剪第二张图像
img2 = load_img(7)
img2_cropped = np.array(img2[1100:1600, 900:1200])

# 创建SIFT检测器
sift = cv.SIFT_create()

# 提取第一张图像的关键点和描述符
keypoints1, descriptors1 = sift.detectAndCompute(img1_cropped, None)

# 提取第二张图像的关键点和描述符
keypoints2, descriptors2 = sift.detectAndCompute(img2_cropped, None)

# 在第一张图像中选择最显著的关键点（具有最大尺度的关键点）
scales1 = [kp.size for kp in keypoints1]
max_scale_index = np.argmax(scales1)
selected_keypoint1 = keypoints1[max_scale_index]
selected_descriptor1 = descriptors1[max_scale_index]

print(f"Selected keypoint in image 1: position ({selected_keypoint1.pt[0]:.1f}, {selected_keypoint1.pt[1]:.1f}), scale {selected_keypoint1.size:.1f}")

# 计算所选关键点的描述符与第二张图像中所有关键点描述符的距离
distances = []
for i, desc2 in enumerate(descriptors2):
    # 计算欧几里得距离
    distance = np.linalg.norm(selected_descriptor1 - desc2)
    distances.append(distance)

# 找到距离最小的关键点
min_distance_index = np.argmin(distances)
matching_keypoint2 = keypoints2[min_distance_index]
min_distance = distances[min_distance_index]

print(f"Best matching keypoint in image 2: position ({matching_keypoint2.pt[0]:.1f}, {matching_keypoint2.pt[1]:.1f}), scale {matching_keypoint2.size:.1f}")
print(f"Distance between descriptors: {min_distance:.2f}")

# 绘制匹配的关键点
# 在第一张图像上绘制选中的关键点
img1_with_keypoint = cv.drawKeypoints(img1_cropped, [selected_keypoint1], None, 
                                     color=(0, 255, 0), flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 在第二张图像上绘制匹配的关键点
img2_with_keypoint = cv.drawKeypoints(img2_cropped, [matching_keypoint2], None, 
                                     color=(0, 255, 0), flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 显示结果
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 显示第一张图像及其选中的关键点
axes[0].imshow(img1_with_keypoint)
axes[0].set_title(f'Image 1 - Selected Keypoint\nPosition: ({selected_keypoint1.pt[0]:.1f}, {selected_keypoint1.pt[1]:.1f})\nScale: {selected_keypoint1.size:.1f}')
axes[0].axis('off')

# 显示第二张图像及其匹配的关键点
axes[1].imshow(img2_with_keypoint)
axes[1].set_title(f'Image 2 - Matching Keypoint\nPosition: ({matching_keypoint2.pt[0]:.1f}, {matching_keypoint2.pt[1]:.1f})\nScale: {matching_keypoint2.size:.1f}')
axes[1].axis('off')

plt.tight_layout()
plt.show()

# 显示匹配质量统计
print(f"\nFeature matching statistics:")
print(f"Total keypoints in image 1: {len(keypoints1)}")
print(f"Total keypoints in image 2: {len(keypoints2)}")
print(f"Descriptor distance: {min_distance:.2f}")
print(f"Position difference: ({abs(selected_keypoint1.pt[0] - matching_keypoint2.pt[0]):.1f}, {abs(selected_keypoint1.pt[1] - matching_keypoint2.pt[1]):.1f})")
print("SIFT feature correspondence analysis completed")
```
